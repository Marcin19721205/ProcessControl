```python
pv_unit = input("Podaj jednostkę PV [Eng] (np. kPa, °C, m): ").strip()  # jednostka PV
out_unit = input("Podaj jednostkę OUT [Eng] (np. %, t/h, kg/h): ").strip()  # jednostka OUT

delta_pv = float(input(f"Podaj ΔPV [{pv_unit}]: "))  # zmiana PV ze step testu
delta_out = float(input(f"Podaj ΔOUT [{out_unit}]: "))  # zmiana OUT ze step testu

pv_min = float(input(f"Podaj PVmin [{pv_unit}]: "))  # dolny zakres PV
pv_max = float(input(f"Podaj PVmax [{pv_unit}]: "))  # górny zakres PV

out_min = float(input(f"Podaj OUTmin [{out_unit}]: "))  # dolny zakres OUT
out_max = float(input(f"Podaj OUTmax [{out_unit}]: "))  # górny zakres OUT

if delta_out == 0:
    raise ValueError("ΔOUT nie może być równe 0.")
if pv_max == pv_min:
    raise ValueError("PVmax i PVmin nie mogą być równe.")
if out_max == out_min:
    raise ValueError("OUTmax i OUTmin nie mogą być równe.")

pv_span = pv_max - pv_min  # zakres PV [Eng]
out_span = out_max - out_min  # zakres OUT [Eng]

kpss_raw = delta_pv / delta_out  # gain procesu surowy [PV_unit/OUT_unit]
kpss_norm = kpss_raw * (100 / pv_span) * (out_span / 100)  # gain procesu znormalizowany [%PV/%OUT]

if kpss_norm == 0:
    raise ValueError("Gain znormalizowany wyszedł 0, więc nie da się policzyć gainu regulatora ani PBand.")

kc_reg = 1 / kpss_norm  # gain regulatora [%OUT/%PV]
pband = 100 / kc_reg  # proportional band [%PV]

delta_pv_from_100_out_eng = kpss_raw * 100  # ile Eng PV zmieni 100% OUT
delta_pv_from_100_out_pct = kpss_norm * 100  # ile %PV zmieni 100% OUT

delta_out_for_100_pv = kc_reg * 100  # ile %OUT potrzeba na 100%PV
delta_out_for_1_pv = kc_reg * 1  # ile %OUT potrzeba na 1%PV

print("\n" + "-" * 90)
print(f"Zakres PV                     = {pv_span:.6f} [{pv_unit}]")
print(f"Opis: pełny zakres pomiarowy PV od PVmin do PVmax.\n")

print(f"Zakres OUT                    = {out_span:.6f} [{out_unit}]")
print(f"Opis: pełny zakres sterowania OUT od OUTmin do OUTmax.\n")

print(f"Kpss surowy                   = {kpss_raw:.6f} [{pv_unit}/{out_unit}]")
print(f"Opis: tyle jednostek PV zmienia się od 1 jednostki OUT.")
print(f"Czyli: 1 {out_unit} zmiany OUT daje {kpss_raw:.6f} {pv_unit} zmiany PV.")
print(f"Czyli: 100 {out_unit} zmiany OUT da {delta_pv_from_100_out_eng:.6f} {pv_unit} zmiany PV.\n")

print(f"Kpss znormalizowany           = {kpss_norm:.6f} [%PV/%OUT]")
print(f"Opis: tyle % zakresu PV zmienia się od 1% zakresu OUT.")
print(f"Czyli: 1% OUT daje {kpss_norm:.6f}% PV.")
print(f"Czyli: 100% OUT da {delta_pv_from_100_out_pct:.6f}% PV.\n")

print(f"Gain regulatora first guess   = {kc_reg:.6f} [%OUT/%PV]")
print(f"Opis: o tyle %OUT zmieni się wyjście regulatora na każdy 1% błędu PV.")
print(f"Czyli: 1% błędu PV daje {delta_out_for_1_pv:.6f}% OUT.")
print(f"Czyli: 100% błędu PV daje {delta_out_for_100_pv:.6f}% OUT.\n")

print(f"PBand                         = {pband:.6f} [%PV]")
print(f"Opis: taki błąd PV powoduje 100% zmiany wyjścia regulatora P.")
print(f"Czyli: 100% OUT odpowiada {pband:.6f}% PV.")
print(f"Czyli: błąd {pband:.6f}% PV wywoła 100% OUT.")
print(f"Czyli: 20% OUT odpowiada {0.2 * pband:.6f}% PV.")
print("-" * 90)
```
