```python
# Skrypt ocenia oscylacje pętli poziomu wg Skogestada i liczy współczynnik korekcyjny f
# Założenie: analizujemy regulator PI/PID z czasem całkowania Ti0 i okresem oscylacji P0

kc0 = float(input("Podaj aktualne Kc regulatora [-]: "))  # bieżące wzmocnienie regulatora
ti0 = float(input("Podaj aktualne Ti0 [s]: "))  # pierwotny czas całkowania
p0 = float(input("Podaj okres oscylacji P0 [s]: "))  # zmierzony okres oscylacji

if kc0 <= 0:
    raise ValueError("Kc musi być > 0.")
if ti0 <= 0:
    raise ValueError("Ti0 musi być > 0.")
if p0 <= 0:
    raise ValueError("P0 musi być > 0.")

graniczne_3ti = 3 * ti0  # granica diagnostyczna Skogestada
stosunek = p0 / ti0  # bezwymiarowy stosunek okresu oscylacji do czasu całkowania
f = 0.1 * (stosunek ** 2)  # współczynnik korekcyjny Skogestada
nowy_iloczyn = f * kc0 * ti0  # nowa wartość iloczynu Kc*Ti po korekcie

# Wariant 1: cała korekta przez Ti
kc_ti_only = kc0  # Kc bez zmian
ti_ti_only = ti0 * f  # Ti zwiększone o faktor f

# Wariant 2: cała korekta przez Kc
kc_kc_only = kc0 * f  # Kc zwiększone o faktor f
ti_kc_only = ti0  # Ti bez zmian

# Wariant 3: korekta rozdzielona po równo geometrycznie
sqrt_f = f ** 0.5  # pierwiastek z f
kc_split = kc0 * sqrt_f  # Kc zwiększone o sqrt(f)
ti_split = ti0 * sqrt_f  # Ti zwiększone o sqrt(f)

if p0 > graniczne_3ti:
    ocena = "WOLNA oscylacja"
    wniosek = "zgodnie ze Skogenstadem zwykle Kc jest za małe"
elif p0 < graniczne_3ti:
    ocena = "SZYBKA oscylacja"
    wniosek = "zgodnie ze Skogenstadem zwykle Kc jest za duże"
else:
    ocena = "GRANICZNA oscylacja"
    wniosek = "P0 = 3*Ti, punkt graniczny diagnostyki"

print("\n" + "=" * 90)
print("OCENA OSCYLACJI PĘTLI POZIOMU")
print("=" * 90)

print(f"Kc0 = {kc0:.6f} [-]")
print("Aktualne wzmocnienie regulatora.\n")

print(f"Ti0 = {ti0:.6f} [s]")
print("Aktualny czas całkowania używany przed korektą.\n")

print(f"P0 = {p0:.6f} [s]")
print("Zmierzony okres oscylacji na pętli poziomu.\n")

print(f"3*Ti0 = {graniczne_3ti:.6f} [s]")
print("Wartość graniczna do diagnostyki rodzaju oscylacji.")
print(f"Porównanie: P0 {'>' if p0 > graniczne_3ti else '<' if p0 < graniczne_3ti else '='} 3*Ti0")
print("----------------------------------")
print("----------------------------------")
print(f"WNIOSEK: {ocena} -> {wniosek}\n")
print("----------------------------------")
print("----------------------------------")

print(f"Stosunek P0/Ti0 = {stosunek:.6f} [-]")
print("Ile razy okres oscylacji jest większy od czasu całkowania.\n")
print("----------------------------------")
print("----------------------------------")

print(f"Współczynnik Skogestada f = 0.1*(P0/Ti0)^2 = {f:.6f} [-]")
print("Tyle razy należy zwiększyć iloczyn Kc*Ti.")
print(f"Czyli: nowy iloczyn (Kc*Ti)_new = {f:.6f} * (Kc*Ti)_old\n")
print("----------------------------------")
print("----------------------------------")
print(f"Stary iloczyn Kc*Ti = {kc0 * ti0:.6f}")
print(f"Nowy  iloczyn Kc*Ti = {nowy_iloczyn:.6f}\n")
print("----------------------------------")
print("----------------------------------")
print("-" * 90)
print("WARIANT 1: korekta tylko przez Ti")
print(f"Kc_new = {kc_ti_only:.6f} [-]")
print(f"Ti_new = {ti_ti_only:.6f} [s]")
print("Zostawiasz Kc bez zmian, całe zwiększenie iloczynu robisz przez Ti.\n")
print("----------------------------------")
print("----------------------------------")
print("WARIANT 2: korekta tylko przez Kc")
print(f"Kc_new = {kc_kc_only:.6f} [-]")
print(f"Ti_new = {ti_kc_only:.6f} [s]")
print("Zostawiasz Ti bez zmian, całe zwiększenie iloczynu robisz przez Kc.\n")
print("----------------------------------")
print("----------------------------------")
print("WARIANT 3: korekta rozdzielona po równo")
print(f"Kc_new = {kc_split:.6f} [-]")
print(f"Ti_new = {ti_split:.6f} [s]")
print("Zwiększasz jednocześnie Kc i Ti o sqrt(f),")
print("tak aby ich iloczyn wzrósł dokładnie o faktor f.\n")
print("----------------------------------")
print("----------------------------------")

print("UWAGA PRAKTYCZNA:")
print("Dla pętli poziomu zwykle bezpieczniej jest najpierw korygować Ti lub rozdzielać korektę,")
print("to są orientacje tylko")
print("=" * 90)
```
